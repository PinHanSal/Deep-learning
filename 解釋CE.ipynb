```python
import matplotlib.pyplot as plt
import math

p1 = 0.7
p2 = 1 - p1

xs = []
ys = []
intervals = 0.001

minp, minv = None, float("inf")
for i in range(998):
    q1 = (i + 1) * intervals
    q2 = 1 - q1
    xs.append(q1)
    ans = p1 * math.log(1/q1, 2) + p2 * math.log(1/q2, 2)
    ys.append(ans)
    if ans < minv:
        minp, minv = q1, ans
print(minp, minv)
plt.plot(xs, ys)
```

```python
p1 = 0.7
p2 = 1 - p1

xs = []
ys = []
intervals = 0.001

minp, minv = None, float("inf")
for i in range(998):
    q1 = (i + 1) * intervals
    q2 = 1 - q1
    xs.append(q1)
    ans = p1 * (1 - q1) + p2 * (1 - q2)
    ys.append(ans)
    if ans < minv:
        minp, minv = q1, ans
print(minp, minv)
plt.plot(xs, ys)
```
